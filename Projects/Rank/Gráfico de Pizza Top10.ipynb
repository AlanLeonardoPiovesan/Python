```python
# biblioteca
import pandas as pd
```

```python
# xlsx e visualização
tabela = pd.read_excel('SD2 ITENS DE VENDA DA NF.xlsx')
display (tabela)
```

```python
vendas_por_produto = tabela.groupby('Produto     ')['Quantidade  '].sum()
```

```python
top_10_produtos = vendas_por_produto.sort_values(ascending=False).head(10)
```

```python
tabela_top_10 = tabela[tabela['Produto     '].isin(top_10_produtos.index)]
```

```python
import matplotlib.pyplot as plt
```

```python
cores = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen', 'lightpink', 'lightblue', 'lightyellow', 'lightcoral', 'lightblue', 'lightgreen']
```

```python
plt.figure(figsize=(8, 8))
plt.pie(top_10_produtos, labels=top_10_produtos.index, autopct='%1.1f%%', startangle=140, colors=cores)
```
